**Create a loyalty flag for existing customers using the transform() and loc() functions**

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(ny.max)
```

```python
df_ords_prods_merge.head(15)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
#Performing multiple aggregations ('min', 'max', 'mean', 'sum')
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'sum']})
```

1. 'Loyal customers' purchase on average cheaper products than 'New customers' and 'Regular customers'
2. Sum of 'Regular customer' is higher as they are more in number than the other two types of customers and also there is no huge difference in mean

**Create a spending flag for each user based on the average price across all their orders using the following criteria**

```python
df_ords_prods_merge['avg_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(ny.mean)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.head(15)
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

**Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column**

```python
df_ords_prods_merge['median_freq'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(ny.median)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_freq'] > 10) & (df_ords_prods_merge['median_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge.head(15)
```

```python
df_ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```
