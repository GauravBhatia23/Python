# Bar Chart

# Line Chart

```python
df_2 = small[['order_hour_of_day','prices']]
```

```python
# line chart for 'prices' and 'order_hour_of_day' columns
line = sns.lineplot(data = df_2, x = 'order_hour_of_day', y = 'prices')
```

**Customer Demographics**

```python
df_3 = small[['Number_of_dependents','age']]
```

```python
# line chart for 'age' and 'Number_of_dependents' columns
line_2 = sns.lineplot(data = df_3, x = 'age', y = 'Number_of_dependents')
```

Number of dependents seems to be similar across the age groups

```python
df_4 = small[['income','age']]
```

```python
#connection between age and income
sctp = sns.scatterplot(x = 'age', y = 'income', data = df_4)
```

Majority of the population do have < $200K income. Income range from $200K and above is more prevalent in 40+ years age group
